```python
dataset = {
    'raw_data':dtarray,
    'modelname':modelname
}
```

```python
table = pd.DataFrame(dataset)
```

```python
인증 = ['CA인증','천식협회인증','아토피안심인증','알레르기협회인증','CADR']
```

```python
for i in 인증:
    table[i]=table['raw_data'].str.contains(i)
```
