```python
i=0
j=1
dtarray=[]
modelname=[]
imgname = []
brand=[]
factory=[]
price=[]
detailurl=[]
detail="http://www.enuri.com/detail.jsp?modelno="
cate='cate=240616'
x=True
```

```python
while True:
        response = requests.get('http://www.enuri.com/lsv2016/ajax/getListGoods_ajax.jsp?random_seq=251&pageNum='+str(j)+'&pageGap=90&tabType=0&cate=240616&keyword=&brand=&factory=&shop_code=&prtmodelno=&gb1=&gb2=&sponsorList=&infoAdList=&IsDeliverySumPrice=N&IsJungoPriceRemove=N&m_price=&start_price=&end_price=',verify=False)
        data=json.loads(response.text)
        print(str(j)+'번째 response')
        i=0
        j+=1
        x=True
        
        if len(data['lpList'])==0:
            print('lpList 값이 0이므로 반복문 종료')
            break
        while x is True:
            try:
                dtarray.append(data['lpList'][i]['spec_vip_tag'])
                modelname.append(data['lpList'][i]['strModelName'])
                imgname.append(data['lpList'][i]['groupModelList'][0]['strSmallImg'])
                brand.append(data['lpList'][i]['strBrand'])
                factory.append(data['lpList'][i]['strFactory'])
                price.append(data['lpList'][i]['lngMinPrice'])
                detailurl.append(detail+str(data['lpList'][i]['intModelNo'])+cate)
                i+=1
                
            except:
                x=False
                print(len(modelname))
        
```

```python
len(modelname)
```

```python
dataset = {
    'raw_data':dtarray,
    'modelname':modelname,
    'imgname':imgname,
    'brand':brand,
    'factory':factory,
    'price':price,
    'detailurl':detailurl
}
```

```python
table = pd.DataFrame(dataset)
```

```python
table
```

```python
인증 = ['CA인증','천식협회인증','아토피안심인증','알레르기협회인증','CADR']
```

```python
for i in 인증:
    table[i]=table['raw_data'].str.contains(i)
```

```python
table['raw_data']=table['raw_data'].map(lambda x: re.sub('\\$TAG\\S|\\$END\\$|_[0-9]+|_|\\<(/)?[a-z]+\\>',' ',x))
```

```python
table['raw_data'].str.split('/',expand=True)
```

```python
# table['방식']=table['raw_data'].str.split('/',expand=True)[1]
방식 = ['음이온식','복합식','물필터식']
for i in 방식:
    table[i]=table['raw_data'].str.contains(i)
table['필터식']=table['raw_data'].str.contains('[^無]필터식')
## 필터식 구분 완료
```

```python
table['사용면적']=table['raw_data'].map(lambda x: x.split('사용면적 :')[-1]).map(lambda x: x.split('/')[0])
```

```python
table['필터종류']=table['raw_data'].map(lambda x: x.split('필터구성:')[-1]).map(lambda x: x.split('/')[0].strip())
```

```python
브랜드 = table['brand'].unique()
제조사 = table['factory'].unique()
```

```python
for i in 브랜드:
    table['modelname']=table['modelname'].map(lambda x: re.sub(i,"",x))
for j in 제조사:
    table['modelname']=table['modelname'].map(lambda x: re.sub(j,"",x).strip())
```

```python
table
```

```python
table['크기']=table['raw_data'].map(lambda x: x.split('크기:')[-1]).map(lambda x: x.split('/')[0])
```

```python
찌꺼기 = table.loc[table['필터종류'].str.contains('필터')==False,:]['필터종류'].unique()
```

```python
for i in 찌꺼기:
    table['사용면적']=table['사용면적'].map(lambda x: re.sub(i,"",x))
    table['필터종류']=table['필터종류'].map(lambda x: re.sub(i,"",x))
    table['크기']=table['크기'].map(lambda x: re.sub(i,"",x))
```
